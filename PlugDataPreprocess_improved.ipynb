```python
import pandas as pd
import numpy as np
import glob
```

```python
#Define path of the Appliances data
commonpath='/Users/HuHaiwei/Desktop/data-x-data/data/House1/01_plugs_csv/Appliance'
pathlist=[]
for i in range(1,8):
    pathlist.append(commonpath+str(i))
pathlist
```

```python
#Define default values for missing files
missing = pd.DataFrame(np.nan, index=range(86400), columns= [0])
missing.head()
missing.columns = ['missing']
missing.head()
```

```python
#Generate time range
import datetime
from datetime import datetime
date = list(pd.date_range(start = '6/1/2012', end='1/23/2013', freq = 'D'))
date = [str(i)[:10] for i in date]
date[:10]
```

```python
#Merge all data into a dataframe & print out the days missing data
import os
house1 = pd.DataFrame()
for path in pathlist:
    namelist = []
    path=os.path.join(path)
    filenames = glob.glob(path +'/*.csv')
    filenames.sort()
    print(len(filenames))
    df=[]
        
    for filename in filenames:
        i=filename[-14:-4]
        namelist.append(i)
   
    for d in date:
        if d in namelist:
            filepath = os.path.join(path,d+'.csv')
            #print('data of', i, 'is not missing')
            getdata = pd.read_csv(filepath,names=[d])
            l = getdata[d].tolist()                   
            df = df+l
        else:
            print('data of', d ,'is missing for', path[-10:])
            df = df + missing['missing'].tolist()

    print(len(df))
    
    colname = path[-10:]
    house1[colname] = df
    

```

```python
timestamp = list(pd.date_range(start='6/1/2012', end='1/24/2013', freq='s'))
timestamp = [str(i)[:19] for i in timestamp]
timestamp = {'DateTime': timestamp}
timestamp = pd.DataFrame(timestamp)
timestamp.head()
```

```python
timestamp.drop(index=(timestamp.shape[0]-1), inplace=True)
house1['DateTime'] = timestamp['DateTime']
house1.set_index('DateTime', inplace=True)
house1.tail()
```

```python
house1.rename(columns={'Appliance1':'Fridge',\
                                'Appliance2':'Dryer',\
                                'Appliance3':'Coffee machine',\
                                'Appliance4':'Kettle',\
                                'Appliance5':'Washing machine',\
                                'Appliane6':'PC',\
                                'Appliance7':'Freezer'}, inplace=True)
```
